```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset = pd.read_csv("/content/electronics.csv")
dataset.head()
```

```python
from datetime import datetime
pd.to_datetime(dataset['timestamp'])
```

```python
dataset['brand'] = dataset['brand'].astype(str)
```

```python
dataset['category'] = dataset['category'].astype(str)
```

```python
dataset['rating']= dataset['rating'].astype(float)
```

```python
dataset['rating'] = dataset['rating'].astype(float)
```

```python
dataset['user_id'] = dataset['user_id'].astype(str)
dataset['item_id'] = dataset['item_id'].astype(str)
dataset.describe()
```

```python
dataset.nunique()
```

```python
dataset['rating'].dropna(inplace = True)
dataset['rating'].drop_duplicates(inplace = True)
dataset.duplicated().sum()
```

```python
dataset.isnull().sum()
```

```python
dataset.isnull().sum()
```

```python
# ratings distribution
sns.countplot(x='rating', data = dataset)
```

```python
#best year of sales
dataset['year'] = pd.DatetimeIndex(dataset['timestamp']).year
dataset.groupby('year')['rating'].count().plot(kind='bar')
```

```python
# brand that sold the most in 2015
dataset_2015 = dataset[dataset['year'] == 2015]
dataset_2015.groupby('brand')['rating'].count().sort_values(ascending=False).head(10).plot(kind ='bar')
```

```python
# product most sold in 2017
dataset_2017 = dataset[dataset['year']==2017]
dataset_2017.groupby('brand')['rating'].count().sort_values(ascending =False).head(10).plot(kind ='bar')
```

```python
#product by category that sold most
dataset.groupby('category')['rating'].count().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
# product that sold least by brand name
dataset.groupby('category')['rating'].count().sort_values(ascending=True).head(10).plot(kind ='pie')
```

```python
# brand percentage sales
dataset.groupby('brand')['rating'].count().sort_values(ascending= True).head(10).plot(kind ='pie')
```
